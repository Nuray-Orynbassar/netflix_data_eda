```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('netflix_titles.csv')
print(df.isnull().sum())
df = df.dropna()
movie = df.loc[(df['type']== 'Movie') & (df['release_year'] > 2018)]
release_year_count = movie['release_year'].value_counts().sort_index()
release_year_count.plot(kind = 'bar', color = 'green')
plt.title("The number of movies over the years(starting with 2019)")
plt.xlabel("Release year")
plt.ylabel("The number of movies")
plt.xticks(rotation=0)
plt.show()
```

```python
movie = df.loc[(df['type']== 'Movie') & (df['release_year'] > 2018)]
country = movie['country'].value_counts().head(20)
country.plot(kind = 'bar', color = 'red')
plt.title("Top 20 countries by number of movies(starting with 2019) ")
plt.xlabel("Country")
plt.ylabel("The number of movies")
plt.xticks(rotation=45,ha = 'right')
plt.show()
```
